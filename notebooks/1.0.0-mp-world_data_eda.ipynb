```python
import seaborn as sns
import matplotlib.pyplot as plt

df_plot = df.copy()
df_plot["country_risk"] = df_plot["country_risk"].astype(int)
order = sorted(df_plot["country_risk"].unique())

fig, ax = plt.subplots(figsize=(10, 6))

sns.countplot(
    data=df_plot,
    x="country_risk",
    hue="country_risk",
    palette="coolwarm",
    order=order,
    legend=False,
    ax=ax,
)

for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f"{height}",
        (p.get_x() + p.get_width() / 2, height),
        ha="center",
        va="bottom",
        fontsize=11,
        fontweight="semibold",
    )

ax.set_title("Number of Countries by Risk Level", fontsize=16, fontweight="bold")
ax.set_xlabel("Country Risk Level", fontsize=13)
ax.set_ylabel("Number of Countries", fontsize=13)
sns.despine(top=True, right=True)
plt.tight_layout()
plt.show()
```
